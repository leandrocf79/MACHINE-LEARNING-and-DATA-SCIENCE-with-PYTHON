# Árvore de decisão na base de crédito

```python
with open ('D:\Arquivos python\MACHINE LEARNING\Bases de dados\credit.pkl', "rb") as f:
    x_credito_treinamento, y_credito_treinamento, x_credito_teste, y_credito_teste = pickle.load(f)
```

```python
x_credito_treinamento.shape, y_credito_treinamento.shape
```

1500 linhas e 3 colunas

```python
x_credito_teste.shape, y_credito_teste.shape
```

```python
# 'entropy' é para manter o padrão das outras aulas
# TREINAMENTO
arvore_credito=DecisionTreeClassifier(criterion ='entropy', random_state=0)
arvore_credito.fit(x_credito_treinamento, y_credito_treinamento)
```

```python
previsores=arvore_credito.predict(x_credito_teste)
previsores
```

```python
# comparar y_credito_teste
y_credito_teste
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_credito_teste, previsores)
```

```python
from yellowbrick.classifier import ConfusionMatrix
cm=ConfusionMatrix(arvore_credito)
cm.fit(x_credito_treinamento, y_credito_treinamento)
cm.score(x_credito_teste, y_credito_teste)
```

```python
print(classification_report(y_credito_teste, previsores))
```

```python
from sklearn import tree
previsores=['income', 'age', 'loan']
fig, axes=plt.subplots(nrows=1, ncols=1, figsize=(20, 20))
tree.plot_tree(arvore_credito, feature_names=previsores, class_names=str(arvore_credito.classes_), filled=True);
fig.savefig('arvore_credito.png')

```

Gerou erro em classes, para corrigir:

```python
from sklearn import tree
previsores=['income', 'age', 'loan']
fig, axes=plt.subplots(nrows=1, ncols=1, figsize=(20, 20))
tree.plot_tree(arvore_credito, feature_names=previsores, class_names=['0', '1'], filled=True);
fig.savefig('arvore_credito.png')
```
